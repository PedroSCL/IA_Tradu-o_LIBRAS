# 🤟 Análise Exploratória — Tradutor de Libras

Notebook de apoio para visualizar os dados coletados e entender o pipeline de IA.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os, sys

sys.path.insert(0, '../src')
from config import CLASSES, DATA_PROCESSED, MODELS_DIR, FIGURES_DIR

plt.style.use('seaborn-v0_8-whitegrid')
print('Configurado!')
```

## 1. Distribuição das Classes

```python
X = np.load(os.path.join(DATA_PROCESSED, 'X.npy'))
y = np.load(os.path.join(DATA_PROCESSED, 'y.npy'))

print(f'Shape de X: {X.shape}')
print(f'Shape de y: {y.shape}')

counts = [(CLASSES[i], int((y==i).sum())) for i in range(len(CLASSES))]
df_counts = pd.DataFrame(counts, columns=['Sinal', 'Amostras'])

fig, ax = plt.subplots(figsize=(8, 4))
sns.barplot(data=df_counts, x='Sinal', y='Amostras', ax=ax, palette='Blues_d')
ax.set_title('Distribuição de Amostras por Classe', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
df_counts
```

## 2. Visualizar uma Sequência de Landmarks

```python
# Exibe a evolução dos primeiros 10 features ao longo dos frames
sample_idx = 0
cls_name   = CLASSES[y[sample_idx]]
seq        = X[sample_idx]  # (frames, 225)

fig, ax = plt.subplots(figsize=(12, 4))
for i in range(10):
    ax.plot(seq[:, i], label=f'feat_{i}', alpha=0.7)
ax.set_title(f'Sequência de Landmarks — Sinal: {cls_name}', fontsize=12)
ax.set_xlabel('Frame')
ax.set_ylabel('Valor do Landmark')
ax.legend(loc='upper right', ncol=5, fontsize=7)
plt.tight_layout()
plt.show()
```
